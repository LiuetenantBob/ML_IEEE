```python
import pandas as pd
import numpy as np
from glob import glob
import cv2
import matplotlib.pylab as plt
```

```python
# carDataset = glob("..input/carDataset/*.jpg")
imMPL = plt.imread("Vroom.jpg")
imCV = cv2.imread("Vroom.jpg")
```

```python
imMPL.shape, imCV.shape
```

```python
imMPL.flatten()
```

```python
pd.Series(imMPL.flatten())
```

```python
# pd.Series(imMPL.flatten()).plot(kind = "hist", bins = 10, title = "Distribution of Pixel Values")
fig, ax = plt.subplots(figsize = (10, 10))
ax.imshow(imMPL)
ax.axis("off")
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize = (15, 5))
axs[0].imshow(imMPL[:,:,0], cmap = "Reds")
axs[0].set_title("Red")
axs[1].imshow(imMPL[:,:,1], cmap = "Greens")
axs[1].set_title("Green")
axs[2].imshow(imMPL[:,:,2], cmap = "Blues")
axs[2].set_title("Blue")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
axs[0].imshow(imMPL)
axs[0].set_title("Matplotlib")
axs[1].imshow(imCV)
axs[1].set_title("CV2")
plt.show()
```

```python
imCV_RGB = cv2.cvtColor(imCV, cv2.COLOR_BGR2RGB)
fig, axs = plt.subplots(1, 2, figsize = (10, 5))
axs[0].imshow(imMPL)
axs[0].set_title("Matplotlib")
axs[1].imshow(imCV_RGB)
axs[1].set_title("CV2")
plt.show()
```

```python
imCV_Gray = cv2.cvtColor(imCV, cv2.COLOR_BGR2GRAY)
fig, ax = plt.subplots(figsize = (10, 10))
ax.imshow(imCV_Gray)
plt.show()
```

```python
imCV_Gray = cv2.cvtColor(imCV, cv2.COLOR_BGR2GRAY)
fig, ax = plt.subplots(figsize = (10, 10))
ax.imshow(imCV_Gray, cmap = "Greys")
plt.show()
```
